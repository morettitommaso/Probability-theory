# La V.C Ipergoemtrica

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom
```

```python
#parameters

N = 20 #numero totale di palline
K = 8 #numero palline che generano un successo
n = 10 #numero di estrazioni
```

```python
#distribuzione ipergeometrica
dist = hypergeom(N,K,n)
```

```python
#valori assumibili
x = np.arange(0, min(K,n)+1) #il +1 perch√® np.arange esclude l'estremo superiore

# PMF e CDF
pmf = dist.pmf(x)
cdf = dist.cdf(x)
```

```python
# Stampa
for k, p in zip(x, pmf):
    print(f"P(X = {k}) = {p:.4f}")
```

```python
#momenti

print("Media:", dist.mean())      
print("Varianza:", dist.var())    
print("Deviazione standard:", dist.std())
print("Asimmetria (skewness):", dist.stats(moments='s'))
print("Curtosi:", dist.stats(moments='k'))
```

```python
# Grafici
plt.figure(figsize=(10, 4))

# PMF
plt.subplot(1, 2, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title("Distribuzione ipergeometrica - PMF")
plt.xlabel("Palline rosse estratte (k)")
plt.ylabel("P(X = k)")
plt.grid(True)

# CDF
plt.subplot(1, 2, 2)
plt.step(x, cdf, where="post")
plt.title("Funzione di ripartizione - CDF")
plt.xlabel("Palline rosse estratte (‚â§ k)")
plt.ylabel("P(X ‚â§ k)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
